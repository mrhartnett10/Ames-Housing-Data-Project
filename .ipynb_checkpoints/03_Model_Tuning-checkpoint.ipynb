### 1. Fitting OLS Model

First let's see how our model performs using our convential OLS (ordinary least squares) more commonly known as linear regression. We've already seen a brief example of this when determining our LINE Assumptions in our earlier notebook, but as a refresher the process is as follows: instantiate the model, fit the model, and evaluate the model. Let's take a look at each step and explain along the way.

**Instantiate**: allows us to create an instance of class and then fit those class variables and methods onto certain variables. This allows us to apply/create our OLS model through fitting it onto our X & y.

```python
#simple enough:
lr = LinearRegression()
```

**Fit:** next is to take that instance and fit it tour training model. Make sure to not fit to our entire X or y or we won't be able to effectively evaluate

```python
#use .fit, make sure to fit onto our scaled data, so Z_train and y_train:
lr.fit(Z_train, y_train)
```

**Evaluate:** So how do we evaluate our model? The metric we are most interested in when evaluating our model is known as the r-squared score (or r^2). What the r^2 score tells us is to what degree our model explains the variability of the intended target. Essentially how "good" our model is. Scores are on a scale of 0 to 1, and the closer to 1 the better our model is at explaining the dataset. From class here's the lectures explanation of r^2 score: 

Another popular regression metric is the $R^2$, or **coefficient of determination**, is the proportion of variability in $y$ we can explain with $x$.

Let 

TSS = $\sum (y_i - \bar{y})^2$ be the **total sum of squares** and

RSS = $\sum (y_i - \hat{y}_i)^2$ be the **residual standard error**.



Then $R^2$ is defined as:

$$R^2 = \frac{\text{TSS - RSS}}{\text{TSS}} = 1 - \frac{\text{RSS}}{\text{TSS}} = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}$$

Typically we will comment our r^2 score in percentage form, so if our score is .802, we can say that our model explains 80.2% of the variability of y ('SalePrice')

Makes sense, now let's get some r^2 scores and see how our model performs! Luckily python allows us to do this a number of ways, the easiest simply using our .score method through our instantiated LinearRegression class. Let's take a look at our two z scores for our train and test set:

```python
print('Train R^2:', lr.score(Z_train, y_train))
print('Test R^2:', lr.score(Z_test, y_test))
```

**Interpretation:** Not bad! Our train set results in a score of roughly .886 and our test performed fairly similarly with a score of .873. We are able to say that we have avoided overfitting to our training set, but can we improve our model somehow? Let's also examine our coefficients for our model. Coefficients are the associated degrees of 'effect' that each variable has on the target. It is important to note that for multilinear regression, the assumption is that everything else is held constant. So if we see a one unit increase in our specific feature variable, we will see a corresponding increase in y by the magnitude of the respective feature coefficient (all else held constant).

### 2. Bias, Variance, and Their Tradeoffs

As we just saw R^2 scores can tell us a lot about our model. Luckily, we were able to achieve respectable scores on our first go around (thanks to effective cleaning, EDA, preprocessing, and feature engineering), but let's say we didn't perform so well. Say we had created a model that resulted in scores of Train: .252 and Test: .232. Clearly not great scores, so what does this tell us? This tells us that our model contains what is known as high bias. Essentially this means that our model is underfit and has done a bad job at predicting the y (compared to the actual y values). 

Let's say for example sake, we scored a Train score of : .724 (not great, but not too bad), and a test score of : .124 (yuck!). How did that happen? What happened was we designed our model that has overfit itself to the training model and is unable to generalize to new data. This is what is known as containing high variance.

The easiest way to understand this is by looking at the following examples from our class lectures:


**B**ias = **B**ad Model, bad at predicting y
<img src="./assets/high_bias_copy.png" style="height: 400px">

**V**ariance = o**V**erly fitted, responds poorly to new data
<img src="./assets/high_variance_copy.png" style="height: 400px">

As you may intuitively gathered each problem arises differently, bias models typically require elements of feature engineering through either more features or creating interaction terms. This allows us to increase the complexity of our model and do a better job at creating accurate predictions. But be careful! There is a tradeoff, as we reduce bias, we begin to introduce more variance. So it is important to understand that relationship as to not tip our model to either extreme. 

So then high variance occurs when we have an overly complex model. This often occurs when we try to include too many features and our model starts receiving too much noise and attempt to account too much for all the variables. A way to remedy this is to decrease the complexity (remove unnecessary features), or the other technique will be to regularize our data which we will explore in the section below.

### 3. Data Regularization: Ridge & LASSO

While our R^2 scores resulted favorly from both a lack of bias and variance standpoint, it is always proper to explore regularizting your data to see if you can improve upon it. Even "good" models can be improved upon. So let's explore this through Ridge Regression.

**Instantiate:** similar to all our other models the first step is to instantiate, the only difference here is that Ridge includes an 'alpha'which as we touched upon earlier is the penalty coefficient introduced to our model. In this case it will be up to us to select the best alpha, but as we will see in a bit we can utilize cross val (CV) to have python determine the 'best' alpha term

```python
#instantiate
ridge_model = Ridge(alpha = 10)
```

**Fit:** next we fit our model, same way we did with our OLS model but using our ridge_model this time

```python
#fitting (don't forget to stay consistent and fit to our scaled data!)
ridge_model.fit(Z_train, y_train)
```

**Evaluate:** Now let's check out the r^2 scores, similar to OLS but with ridge again

```python
print('Train R^2:', ridge_model.score(Z_train, y_train))
print('Test R^2:', ridge_model.score(Z_test, y_test))
```

Now let's compare that our earlier OLS scores: 

```python
print('OLS Train R^2:', lr.score(Z_train, y_train))
print('OLS Test R^2:', lr.score(Z_test, y_test))

print('Ridge Train R^2:', ridge_model.score(Z_train, y_train))
print('Ridge Test R^2:', ridge_model.score(Z_test, y_test))
```

As we can see Ridge resulted in slightly lower R^2 score for train, but a higher R^2 score for our test set. So we have reduced a degree of overfitting, but have arguably introduced slightly more bias in our train model. As noted before this is due to the bias-variance tradeoff and will be up to us to determine the sweet spot. 

Now as I mentioned above about the alpha, what happens if we change that? Is 10 the best alpha we can use? Well there's a method involving cross validation AND Ridge with what is known as RidgeCV. As we will see in a second, RidgeCV aims to leverage cross validation and provide us with the best alpha coefficient. For this we will create a list of values for our alphas using np.logspace (creates equidistant values given a set of parameters, and then uses that value as an exponent for base 10). We will then set our cv parameter to 5 in this case to cross validate over the list of potential alphas to find the best one for our model. Finally we can fit it and test out the scores again to see if there's been any improvement!

```python
#Set up alphas with logspace technique, as discussed in class this creates 100 equidistant values between 0 and 5
#and then takes 10^that number
r_alphas = np.logspace(0, 5, 100)
```

```python
#Next we cross validate:
ridge_cv = RidgeCV(alphas = r_alphas, scoring = 'r2', cv = 5)
```

```python
#Now we can fit it to our model again:
ridge_cv = ridge_cv.fit(Z_train, y_train)
```

```python
#We can also check to see what alpha was selected by our model during the cross val process:
ridge_cv.alpha_
```

**Evaluate:** Now we are ready to evaluate again and check our r^2 scores, let's see if there's any improvement

```python
print('Ridge CV Train R^2:', ridge_cv.score(Z_train, y_train))
print('Ridge CV Test R^2:', ridge_cv.score(Z_test, y_test))
```

```python
#Now let's compare all 3:
print('OLS Train R^2:', lr.score(Z_train, y_train))
print('OLS Test R^2:', lr.score(Z_test, y_test))

print('Ridge Train R^2:', ridge_model.score(Z_train, y_train))
print('Ridge Test R^2:', ridge_model.score(Z_test, y_test))

print('Ridge CV Train R^2:', ridge_cv.score(Z_train, y_train))
print('Ridge CV Test R^2:', ridge_cv.score(Z_test, y_test))
```

We can see that with Ridge CV, we have increased our Train score compared to regular ridge, but still slightly less than OLS, but our test Ridge CV is higher than our normal Ridge. While the difference is slightly bigger, again it is up to us to evaluate which model we think will be optimal.

Let's check out Lasso regularization method and see how that does. As we briefly noted, Lasso performs essentially the same feat as Ridge by introducing a penalty term, but incorporates feature selection by also penalizing the feature coefficients. Ineffective or unnecessary features will have coefficients drift towards 0 and be phased out of the model

**Instantiate:** Continuing the trend, let's instantiate our Lasso model

```python
lasso_model = Lasso(alpha = 10)
```

**Fit:** Now we fit the model (notice a pattern here??)

```python
lasso_model.fit(Z_train, y_train)
```

**Let Me Guess...Evaluate?** Bingo, let's check the r^2 scores

```python
print('Lasso Train R^2:', lasso_model.score(Z_train, y_train))
print('Lasso Test R^2:', lasso_model.score(Z_test, y_test))
```

Before comparing back to the other scores, let's perform the Lasso Cross Val method. It's similar steps to RidgeCV in that we input different parameters. Let's use the example we saw in our class lecture:

```python
# Set up a list of Lasso alphas to check.
l_alphas = np.logspace (0, 3, 100)

# Cross-validate over our list of Lasso alphas.
lasso_cv = LassoCV(alphas = l_alphas, cv = 5, max_iter = 50000)

# Fit model using best ridge alpha!
lasso_cv.fit(Z_train, y_train);
```

Let's check out the R^2 scores

```python
print('LassoCV Train R^2:', lasso_cv.score(Z_train, y_train))
print('LassoCV Test R^2:', lasso_cv.score(Z_test, y_test))
```

Okay, now let's put them all together in a neat little box and take a look at our r^2 scores:

```python
print(" OLS ".center(18, "="))
print('OLS Train R^2:', lr.score(Z_train, y_train))
print('OLS Test R^2:', lr.score(Z_test, y_test))
print()
print(" Ridge ".center(18, "="))
print('Ridge Train R^2:', ridge_model.score(Z_train, y_train))
print('Ridge Test R^2:', ridge_model.score(Z_test, y_test))
print()
print(" RidgeCV ".center(18, "="))
print('Ridge CV Train R^2:', ridge_cv.score(Z_train, y_train))
print('Ridge CV Test R^2:', ridge_cv.score(Z_test, y_test))
print()
print(" Lasso ".center(18, "="))
print('Lasso Train R^2:', lasso_model.score(Z_train, y_train))
print('Lasso Test R^2:', lasso_model.score(Z_test, y_test))
print()
print(" LassoCV ".center(18, "="))
print('LassoCV Train R^2:', lasso_cv.score(Z_train, y_train))
print('LassoCV Test R^2:', lasso_cv.score(Z_test, y_test))
```

Wow, all very close together! Luckily we didn't need much regularization required since we started with a model that was low in variance to begin (proper feature selection and preprocessing). So which version of these should we use for our model then? Well, it's up to us to decide where our happy medium lies. As we can see our train score is highest in our regular OLS model, but the difference between train and test is the largest at magnitude of .0127, so you could argue is the most overfit. The model with the least 'overfittedness' is our regular Ridge with alpha at 10, with a difference of .0062, but the model I believe to be the "happiest medium" is the RidgeCV model with a difference of .0098, but performed the best in our train model in terms of R^2 score (amongst our regularization attempts).
